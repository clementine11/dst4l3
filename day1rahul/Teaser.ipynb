##Vectorization

The python numerical stack is based on a concept called vectorization. You have seen this in spreadsheets where you manipulate an entire column.

Whats the idea? It is that cpu's have some ability to carry out multiple things at one time, and we should use it to increase speed. But its also something else: there is a simplicity of notation. For example, see the following

````python
a=[1,2,3,4,5]
b=[1,2,3,4,5]
print type(a)
print "============="
for index,aitem in enumerate(a):
    print index, a[index], b[index], a[index]*b[index]
````

>indexing starts at zero

>doing something with an entire column is called vectorization

````python
a=np.array([1,2,3,4,5])
print type(a)
b=np.array([1,2,3,4,5])

print a*b
````

````python
sns.set_context("notebook") #different formats, e.g., paper, poster
meanrat=df.rating.mean()
print meanrat, np.mean(df.rating), df.rating.median()
with sns.axes_style("whitegrid"): #alternatives: white, ticks, dark, darkgrid
    df.rating.hist(bins=30, alpha=0.4); #alpha transparency
    plt.axvline(meanrat, 0, 0.75, color='r', label='Mean') #mean line goes .75 up to top. 1.0 would go to top of graph
    plt.xlabel("average rating of book")
    plt.ylabel("Counts")
    plt.title("Ratings Histogram")
    plt.legend()
    #sns.despine() can take away spines 
````

````python
sns.set_context("poster") 
meanrat=df.rating.mean()
print meanrat, np.mean(df.rating), df.rating.median()
with sns.axes_style("whitegrid"): 
    df.rating.hist(bins=30, alpha=0.4); 
    plt.axvline(meanrat, 0, 1.0, color='r', label='Mean')     
    plt.xlabel("average rating of book")
    plt.ylabel("Counts")
    plt.title("Ratings Histogram")
    plt.legend()
    #sns.despine() can take away spines 
````
